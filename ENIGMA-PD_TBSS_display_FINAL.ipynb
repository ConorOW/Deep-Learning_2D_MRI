# White matter abnormalities in Parkinson disease: An ENIGMA-PD TBSS Study

This is a notebook that shows how our different DTI metrics vary across sites. We will filter the dataframes just to focus on PD subjects.

### Notebook Setup

```python
# Import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
import ptitprince as pt
import matplotlib.patches as patches
```

### Data

```python
# Load in our dataset
df = pd.read_csv('./full_dataframe_OLD.csv', header=0, index_col='SubjID')

# Drop rows where age, sex or diagnosis information is missing
df = df.dropna(subset=['diagnosis', 'sex', 'age'])

# Swap out the 0s and 1s in diagnosis
df['diagnosis'] = df['diagnosis'].replace({0: 'Control', 1: 'PD'})

# Let's filter our dataframe to just include PD subjects
df = df[df['diagnosis'] == 'PD'] 

# Import our dMRI site parameters
# params = pd.read_csv('./ENIGMA-PD_dMRI_data.csv', header=0, index_col="Site")
```

### Take a look at the data

```python
# Get some information on our dataframe
subjects, variables = df.shape

print(f"Number of Parkinson disease participants: {subjects}")
print(f"Number of variables: {variables}")
```

### Age and Whole-brain Average FA
First we will look at the relationship between whole-brain Average FA and participant age

```python
# Let's do some plotting
plt.figure(figsize=(10,8))

# Main figure
sns.lmplot(x="age", 
           y="Average_FA", 
           data=df, 
           fit_reg=False, 
           hue='site',
           height=10,
           legend=False)

plt.legend(bbox_to_anchor=(1.04, 1), 
           fontsize='x-large', 
           title_fontsize='32',
           loc="upper left")

# Add a title
plt.title('Age and Whole-brain Average FA \n across different ENIGMA-PD Sites', 
          fontsize=25, 
          fontweight='bold')

# Set a label for our x-axis
plt.xlabel("Age",
            fontweight ='bold',
            size=18)

# Set a label for our y-axis
plt.ylabel("Whole-brain Average FA",
           fontweight ='bold',
           size=18)

# Add a grid
plt.grid(color='lightgray', 
         alpha=0.5, 
         zorder=1)

# Create a Rectangle patch
rect = patches.Rectangle((50, 0.27), 40, 0.13, linewidth=2, edgecolor='r', facecolor='none')

# Add the patch to the Axes
plt.gca().add_patch(rect)

# Set tight layout
plt.tight_layout()

# Save figure
# plt.savefig('./Average_FA_age.png', dpi=300, facecolor='white')

# Display our figure
plt.show()
```

From what we can see above, there look to be PD participants from Manchester and the Taiwan site who fall well outside of
the normal bounds for other sites, but beyond this, things look ok.

### Violin Plots

```python
plt.figure(figsize=(12,12))

plt.title(label="Average FA Across ENIGMA-PD Sites", fontsize=20)

sns.stripplot(y="Average_FA", 
              x="site", 
              data=df,
              linewidth=2,
              size=10)

pt.half_violinplot(x = "site",
                   y = "Average_FA",
                   data = df,
                   width = .9,
                   gridsize = 300)

plt.xlabel(xlabel="Site", fontsize=18)
plt.ylabel(ylabel="Whole-brain Average FA", fontsize=18)
plt.xticks(rotation=80, fontsize=14)
plt.yticks(fontsize=14)

# Save figure
plt.savefig('./Average_FA.png', dpi=300, facecolor='white')

plt.show()
```
